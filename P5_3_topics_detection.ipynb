```python
import os
import pandas as pd
import glob
import numpy as np
import pickle
import gensim
import matplotlib.pyplot as plt

from gensim.models.nmf import Nmf
import pyLDAvis
import pyLDAvis.gensim
pyLDAvis.enable_notebook()

import warnings
warnings.simplefilter("ignore")
warnings.warn("deprecated", DeprecationWarning)
warnings.simplefilter("ignore")
```

# Import corpus (version Large)

```python
# Post
df_post_train = pd.read_pickle('train/X_train_filtre.pkl')
df_post_train = df_post_train.drop(columns=['Text', 'text_large'])
df_post_train = df_post_train.rename(columns={'text_small': 'post'})
df_post_train.head(5)
```

```python
df_post_train.shape
```

```python
# build dictionary
dictionary = gensim.corpora.Dictionary(df_post_train['post'])
print('Dictionnaire : {} mots'.format(len(dictionary)))
```

```python
# build bow
bow = [dictionary.doc2bow(text) for text in df_post_train['post'].values]
```

# Latent Dirichlet Allocation (Gensim)

## Choix du nombre de topics à détecter

```python
def lda(num_topics):
    # model
    lda_model = gensim.models.ldamodel.LdaModel(corpus=bow,
                                                id2word=dictionary,
                                                num_topics=num_topics,
                                                alpha='asymmetric',
                                                random_state=1,
                                                # workers=None, # 3
                                                per_word_topics=False)
    # coherence Score
    coherence_model_lda = gensim.models.CoherenceModel(model=lda_model,
                                                       texts=df_post_train['post'],
                                                       dictionary=dictionary,
                                                       coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    return coherence_lda, lda_model
```

```python
# compute lda and coherence score for different numbers of topics
num_topics = list(np.arange(2, 50, 2))
coherences = []
for num in num_topics:
    score, _ = lda(num)
    coherences.append(score)
    print('Nombre de topics : {}, coherence : {}'.format(num, score))
```

```python
# plot coherence score
plt.plot(num_topics, coherences)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.show()
```

A partir du graph ci-dessus, nous allons retenir 12 topics car le maximum est atteint à 12 topics.

## Calcul LDA

```python
num_topic = 12
_, lda_model = lda(num_topic)
```

```python
lda_model.save('model/lda_model_{}'.format(num_topic))
```

```python
#load model
num_topic = 12
lda_model = gensim.models.ldamodel.LdaModel.load('model/lda_model_12')
```

```python
# get topics matrix
mat = lda_model.get_topics()
# save matrix
with open('PickleData/lda_matrix_{}.pkl'.format(num_topic), 'wb') as f:
    pickle.dump(mat, f, pickle.HIGHEST_PROTOCOL)
```

```python
# get features
def get_lda_features(post):
    post_bow = dictionary.doc2bow(post)
    features = lda_model.get_document_topics(post_bow, minimum_probability=0.)
    features = gensim.matutils.sparse2full(features, num_topic)
    return features


df_features = df_post_train['post'].apply(func=get_lda_features)
df_result = df_features.apply(pd.Series)

# concatenate post and features
df_post_train = pd.concat([df_post_train, df_result], axis=1)
df_post_train.head(5)
```

```python
# save features
df_post_train.to_pickle('PickleData/lda_features_{}.pkl'.format(num_topic))
```

## Visualisation des topics

```python
# Visualize the topics
vis = pyLDAvis.gensim.prepare(lda_model, bow, dictionary)
vis
```

```python
def get_topics(model, num_topics):
    word_dict = {}
    for i in range(num_topics):
        words = model.show_topic(i, topn=20)
        word_dict['Topic # ' + '{:02d}'.format(i+1)] = [i[0] for i in words]
    return pd.DataFrame(word_dict)
```

```python
get_topics(lda_model, num_topic)
```

# Non-negative matrix factorization

## Choix du nombre de topics à détecter

```python
def nmf(num_topics):
    # model
    nfm_model = Nmf(corpus=bow,
                    id2word=dictionary,
                    num_topics=num_topics,
                    random_state=1)
    # coherence Score
    coherence_model = gensim.models.CoherenceModel(model=nfm_model,
                                                   texts=df_post_train['post'],
                                                   dictionary=dictionary,
                                                   coherence='c_v')
    coherence = coherence_model.get_coherence()
    return coherence, nfm_model
```

```python
num_topics = list(np.arange(2, 50, 2))
coherences = []
for num in num_topics:
    score, _ = nmf(num)
    coherences.append(score)
    print('Nombre de topics : {}, coherence : {}'.format(num, score))
```

```python
# plot coherence score
plt.plot(num_topics, coherences)
plt.xlabel("Num Topics")
plt.ylabel("Coherence score")
plt.show()
```

Nous allons choisir de retenir 18 topics pour notre étude.

## Calcul NMF

```python
num_topic = 18
_, nmf_model = nmf(num_topic)
get_topics(nmf_model, num_topic)
```

```python
# Save model
nmf_model.save('model/nmf_model_{}'.format(num_topic))
```

```python
# get features
def get_nmf_features(post):
    post_bow = dictionary.doc2bow(post)
    features = nmf_model.get_document_topics(post_bow, minimum_probability=0.)
    features = gensim.matutils.sparse2full(features, num_topic)
    return features


df_features = df_post_train['post'].apply(func=get_nmf_features)
df_result = df_features.apply(pd.Series)

# concatenate post and features
df_post_train = pd.concat([df_post_train[['post']], df_result], axis=1)
df_post_train.head(5)
```

```python
# save features
df_post_train.to_pickle('PickleData/nmf_features_{}.pkl'.format(num_topic))
```
